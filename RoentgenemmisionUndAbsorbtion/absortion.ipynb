```python
#Bestimmung der Rydbergkonstante



#Daten zu Liste zusammenfassen
EK=[EBrom,EZirkonium,EZink,EStrontium,ERubidium,EGallium]
WEK=np.sqrt(EK)
print(EK,WEK)
Ordzahl=[35,40,30,38,37,31]
SigmaO=[SigmaBrom,SigmaZirkonium,SigmaZink,SigmaStrontium,SigmaRubidium,SigmaGallium]
Ordzah=[0,0,0,0,0,0]
xyz=0
while xyz < 6:
    Ordzah[xyz]=Ordzahl[xyz]-SigmaO[xyz]
    xyz=xyz+1
#Fit an Messpunkte
params, covariance_matrix = np.polyfit( Ordzah,WEK, deg=1, cov=True)

errors = np.sqrt(np.diag(covariance_matrix))

for name, value, error in zip('ab', params, errors):
    print(f'{name} = {value:.3f} ± {error:.3f}')
#Plot
x_plot=np.linspace(26,36)
plt.plot(x_plot,params[0] * x_plot + params[1],label='Lineare Regression')
plt.plot(Ordzah,WEK,'ro')
plt.ylabel('$\sqrt{E_K}$')
plt.xlabel('Ordnungszahl (Z-$\sigma$)')
plt.title('Rydbergkonstante')
plt.legend(loc='best')
plt.savefig('rydberg.pdf')
print(params[0]*const.h)
```
