```python
#Berechnung Wrmestrom
x=0.03 #Abstand der Thermoelemente
#Literaturwerte fr Kappa
KMessing=120
KAluminium=236
KEdelstahl=30
# Werte fr die Querschnittsflche aus der Anleitung
AMessingb=0.012*0.004
AMessings=0.012*0.004
AAluminum=0.007*0.004
AEdelstahl=0.012*0.004
#Zeitpunkte
z1=int(len(T11)/5)
z2=int(2*len(T11)/5)
z3=int(3*len(T11)/5)
z4=int(4*len(T11)/5)
z5=len(T11)-1

print('Zeitpunkte:',z1/5,z2/5,z3/5,z4/5,z5/5)
# Berechnung
QMb1=-1*KMessing*AMessingb*(T12[z1]-T11[z1])/3
QMb2=-1*KMessing*AMessingb*(T12[z2]-T11[z2])/3
QMb3=-1*KMessing*AMessingb*(T12[z3]-T11[z3])/3
QMb4=-1*KMessing*AMessingb*(T12[z4]-T11[z4])/3
QMb5=-1*KMessing*AMessingb*(T12[z5]-T11[z5])/3

QMs1=-1*KMessing*AMessings*(T13[z1]-T14[z1])/3
QMs2=-1*KMessing*AMessings*(T13[z2]-T14[z2])/3
QMs3=-1*KMessing*AMessings*(T13[z3]-T14[z3])/3
QMs4=-1*KMessing*AMessings*(T13[z4]-T14[z4])/3
QMs5=-1*KMessing*AMessings*(T13[z5]-T14[z5])/3

QA1=-1*KAluminium*AAluminum*(T16[z1]-T15[z1])/3
QA2=-1*KAluminium*AAluminum*(T16[z2]-T15[z2])/3
QA3=-1*KAluminium*AAluminum*(T16[z3]-T15[z3])/3
QA4=-1*KAluminium*AAluminum*(T16[z4]-T15[z4])/3
QA5=-1*KAluminium*AAluminum*(T16[z5]-T15[z5])/3

QE1=-1*KEdelstahl*AEdelstahl*(T17[z1]-T18[z1])/3
QE2=-1*KEdelstahl*AEdelstahl*(T17[z2]-T18[z2])/3
QE3=-1*KEdelstahl*AEdelstahl*(T17[z3]-T18[z3])/3
QE4=-1*KEdelstahl*AEdelstahl*(T17[z4]-T18[z4])/3
QE5=-1*KEdelstahl*AEdelstahl*(T17[z5]-T18[z5])/3

QMb=[QMb1,QMb2,QMb3,QMb4,QMb5]
QMs=[QMs1,QMs2,QMs3,QMs4,QMs5]
QA=[QA1,QA2,QA3,QA4,QA5]
QE=[QE1,QE2,QE3,QE4,QE5]

#Durchschnitt berechnen
QMBd=sum(QMb)/len(QMb)
QMBdf=np.std(QMb,ddof=1)/np.sqrt(np.size(QMb))
QMBfb=ufloat(QMBd,QMBdf)

QMsd=sum(QMs)/len(QMs)
QMsdf=np.std(QMs,ddof=1)/np.sqrt(np.size(QMs))
QMsfb=ufloat(QMsd,QMsdf)

QAd=sum(QA)/len(QA)
QAdf=np.std(QA,ddof=1)/np.sqrt(np.size(QA))
QAfb=ufloat(QAd,QAdf)

QEd=sum(QE)/len(QE)
QEdf=np.std(QE,ddof=1)/np.sqrt(np.size(QE))
QEfb=ufloat(QEd,QEdf)

print('Wrmestrme Messing breit',QMb)
print('Wrmestrme Messing schmal',QMs)
print('Wrmestrme Aluminium',QA)
print('Wrmestrme Edelstahl',QE)

print('Durchschnitt Mb=',QMBfb)
print('Durchschnitt Ms=',QMsfb)
print('Durchschnitt Alu=',QAfb)
print('Durchschnitt Ede=',QEfb)
```
