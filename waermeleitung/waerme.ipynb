```python
def peakfinder(L,f=1,ug=0,og=1):
    lauf=ug
    index=0
    lange=0
    Z=[0]
    zwischenspeicher=Z*og
    i=0
    j=0    
    while lauf < og:
        if L[lauf] > L[lauf-1] and L[lauf] > L[lauf+1] :
            zwischenspeicher[index]=lauf
            index=index+1
            lauf=lauf+f
        elif L[lauf]==L[lauf-1]and L[lauf-1] > L[lauf-2]:
            run=0
            mittel=0
            while L[lauf]==L[lauf-1] :
                run=run+1
                mittel=mittel+lauf
                lauf=lauf+1
            zwischenspeicher[index]=round(mittel/run)
            index=index+1
            lauf=lauf+f
        else:
            lauf=lauf+1
        
    while i < len(zwischenspeicher):
        if zwischenspeicher[i] != 0:
            lange=lange+1
            i=i+1
        else:
            i=i+1
    indexes=Z*lange
    while j < lange:
        indexes[j]=zwischenspeicher[j]
        j=j+1
    return indexes
```

```python
def umschreiben(Daten,indexliste):
    Z=[0]
    i=0
    rueckgabe=Z*len(indexliste)
    while i < len(indexliste):
        rueckgabe[i]=Daten[indexliste[i]]
        i=i+1
    return rueckgabe
```
